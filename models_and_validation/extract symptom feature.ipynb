### The corresponding 322-dimensional symptom vectors are extracted for 4219 diseases, and there is no correlation to fill 0

```python
dis_symfea={}
for d in list(lis_dis.index):
    d_fea=[]
    for s in list(lis_sym.index):
        for i in range(len(sym)):
            if sym.at[i,"MeSH Disease Term"] ==sym.at[i,"MeSH Symptom Term"]:
                d_fea.append(sym.at[i,"TFIDF score"])
                break
        d_fea.append(0)
    print(s)
    dis_symfea[d] = d_fea
```

```python
result = pd.DataFrame(dis_symfea)
#shape=(322, 4220)
#result.to_csv("result.csv")
dis_fea = result.T
#shape=(4219, 323)
#dis_fea.to_csv("dis_fea.csv")
```
