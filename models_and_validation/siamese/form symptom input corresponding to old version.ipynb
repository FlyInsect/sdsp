```python
import numpy as np
import pandas as pd
```

```python
dis_fea = pd.read_csv("../dis_fea.csv")
dis_fea["Unnamed: 0"] = dis_fea["Unnamed: 0"].str.lower()
dis_fea.head()
```

### Constuct input feature mat of disease pairs

```python
dis_simWithfea = pd.read_csv("../dis_simWithfea.csv")
print(dis_simWithfea.shape)
dis_simWithfea.head()
```

```python
dis_simWithfea.tail()
```

```python
disA_fea_mat = pd.DataFrame()
disB_fea_mat = pd.DataFrame()

for i in range(len(dis_simWithfea)):
    disA = dis_simWithfea.loc[i,"disease_A"]
    disB = dis_simWithfea.loc[i,"disease_B"]
    print(disA,"\t",disB)
    
    disA_fea = dis_fea.loc[dis_fea["Unnamed: 0"]==disA]
    disB_fea = dis_fea.loc[dis_fea["Unnamed: 0"]==disB]
    
    disA_fea_mat=pd.concat([disA_fea_mat,disA_fea.iloc[:,1:]],axis=0)
    disB_fea_mat=pd.concat([disB_fea_mat,disB_fea.iloc[:,1:]],axis=0)
```
