```python
import numpy as np
import pandas as pd
```

```python
ori_matA = pd.read_csv("../disA_fea_mat.csv",index_col=0)
ori_matA.head()
```

```python
ori_matB = pd.read_csv("../disB_fea_mat.csv",index_col=0)
print(ori_matB.shape)
```

```python

print(max(ori_matA.index))
print(len(ori_matA.index.unique()))
```

```python
muti_label = pd.read_csv("../Mesh_files/label_array.csv",index_col=0)
print(muti_label.shape)
muti_label.head()
```

```python
dis_simWithfea = pd.read_csv("../../dis_simWithfea.csv")
print(dis_simWithfea.shape)
dis_simWithfea.head()
```

```python
muti_label_A = pd.DataFrame()
muti_label_B = pd.DataFrame()

for dis in dis_simWithfea["disease_A"]:
    if(dis in ['sleep disorders', 'metabolic syndrome x']):
        muti_label_A=muti_label_A.append(pd.Series([0 for i in range(24)],name=dis,index=muti_label.columns))
    else:
        muti_label_A=muti_label_A.append(muti_label.loc[dis])
    
for dis in dis_simWithfea["disease_B"]:
    if(dis in ['sleep disorders', 'metabolic syndrome x']):
        muti_label_B=muti_label_B.append(pd.Series([0 for i in range(24)],name=dis,index=muti_label.columns))
    else:
        muti_label_B=muti_label_B.append(muti_label.loc[dis])

print(muti_label_A.shape,muti_label_B.shape)
```

```python
# muti_label_A.to_csv("mesh_label_A.csv")
# muti_label_B.to_csv("mesh_label_B.csv")
```

```python
muti_label_A.head()
```

```python
concat_feaA = pd.concat([ori_matA.reset_index(drop=True),muti_label_A.reset_index(drop=True)],axis=1,ignore_index=True)
print(concat_feaA.shape) # 322+24=346
concat_feaA.head()
```

```python
# concat_feaA.to_csv("con_feaA.csv")
```

```python
concat_feaB = pd.concat([ori_matB.reset_index(drop=True),muti_label_B.reset_index(drop=True)],axis=1,ignore_index=True)
print(concat_feaB.shape) # 322+24=346
# concat_feaB.to_csv("con_feaB.csv")
```

###  concatenate fea_matA,fea_matB

```python
concatAB_fea = pd.concat([concat_feaA,concat_feaB],axis=1,ignore_index=True)
print(concatAB_fea.shape)
concatAB_fea.head()
```
