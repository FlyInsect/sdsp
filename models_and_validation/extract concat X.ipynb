### Construct the input of the ML model and concatenate the symptom vectors of the disease pairs

```python
dis_mat = pd.DataFrame()
for i in range(len(y)):
    disA = y.loc[i,"disease_A"]
    disB = y.loc[i,"disease_B"]
    
    disA_fea = dis_fea.loc[dis_fea["Unnamed: 0"]==disA]
    disB_fea = dis_fea.loc[dis_fea["Unnamed: 0"]==disB]
    if len(disA_fea)==0:
        print(disA)
        continue
    if len(disB_fea)==0:
        print(disB)
        continue
    
    #reset index
    disA_fea.index = [i]
    disB_fea.index = [i]
    disAB_fea = pd.concat([disA_fea.iloc[:,1:],disB_fea.iloc[:,1:]],axis = 1)
    #print(disAB_fea)
    dis_mat = dis_mat.append(disAB_fea)
```
