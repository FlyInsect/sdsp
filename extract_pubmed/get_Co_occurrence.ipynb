```python
"""Count the number of co-occurrence literatures of disease J and symptom I
dis:List of disease related literature PMID
sym:List of symptom....
"""
@jit(parallel=True)
def count_Co_occur(dis,sym):
    count = 0
    for symptom in sym:
        if symptom in dis:
            count+=1
    return count
```

```python
dis_count_dic = {}
count=0
for dis_file in os.listdir("dis_PMIDlist"):
    if dis_file==".ipynb_checkpoints":
        continue
        
    dis_df = pd.read_csv("./dis_PMIDlist/{}".format(dis_file),index_col=0).iloc[:,0]
    dis_count_ls =[]
    sym_name_ls = []

    dis_ls = list(dis_df)
    for sym_file in os.listdir("sym_PMIDlist"):
        sym_df = pd.read_csv("./sym_PMIDlist/{}".format(sym_file),index_col=0).iloc[:,0]
        
        sym_ls = list(sym_df)
        co_count = count_Co_occur(dis_ls,sym_ls)
        dis_count_ls.append(co_count)
        
        sym_name_ls.append(sym_file.split(".")[0])
    dis_count_dic[dis_file.split(".")[0]]=dis_count_ls
    print(count+=1)
    
dis_count_df = pd.DataFrame(dis_count_dic,index=sym_name_ls)
print(dis_count_df.shape)

# dis_count_df.to_csv("Feature2020.csv")
```
