### Extracting up- and down-regulated gene set data

```python
dis_DEG = pd.DataFrame()
for i in range(len(load_dict)):
    tem_dic = {}
    
    down_genes = load_dict.iloc[i,8] #down_genes
    down_len = len(down_genes)
    up_genes = load_dict.iloc[i,9]  #up_genes
    up_len = len(up_genes)
    
    for j in range(down_len):
        tem_dic[down_genes[j][0]]=down_genes[j][1]
    for k in range(up_len):
        tem_dic[up_genes[k][0]]=up_genes[k][1]
    if not pd.isnull(load_dict.iloc[i,3]):         
        genes_df = pd.DataFrame(tem_dic,index=[load_dict.iloc[i,3]])  #disease_UMLS-CUI,include Nan
        #genes_df = pd.DataFrame(tem_dic,index=[load_dict.iloc[i,10]])  #disease_name
        dis_DEG = dis_DEG.append(genes_df)
```

### Multiple signatures of the same disease, average up and down genes

```python
dis_dup ={}

for dis in dis_DEG.index:
    if dis not in dis_dup.keys():
        dis_dup[dis] =1 
    else:
        dis_dup[dis]+=1

print(len(dis_dup.keys())) #223
dis_dup
```

```python
dis_DEG = dis_DEG.fillna(0)
print(dis_DEG.shape)
dis_DEG.head()
```

```python
ave_dis_DEG = pd.DataFrame(columns=dis_DEG.columns)

for dis in dis_dup.keys():
    tem_df = dis_DEG[dis_DEG.index==dis]
    print(dis)
    if dis_dup[dis] > 1:
        ave_dis_DEG=ave_dis_DEG.append(pd.Series(tem_df.mean(axis=0),name=dis),ignore_index=False)
        
    elif dis_dup[dis] ==1:
        ave_dis_DEG=ave_dis_DEG.append(tem_df,ignore_index=False)
    else:
        print("遍历疾病错误")
        
print(ave_dis_DEG.shape)
ave_dis_DEG.head()
```

### Count the occurrences of each gene in all diseases
#### <font color=red>Take the number of occurrences as a threshold</font>，Screening genes<br>
### The size of the feature matrix----name as index：（226，19717），CUI as index：（223，19563）

```python
genes_dup ={}

for gene in dis_DEG.columns:
    genes_dup[gene] = 0
    for occ in list(ave_dis_DEG[gene]):
        if occ != 0:
            genes_dup[gene] +=1

genes_dup
```

```python
print(len(genes_dup))
```

```python
count = 0
count_threshold = 0
_max= 0 
fea_genes=[]

for gene in genes_dup:
    if genes_dup[gene]==0:
        count+=1
    if genes_dup[gene]>=(100):
        count_threshold+=1
        fea_genes.append(gene)
    if genes_dup[gene]>_max:
        _max = genes_dup[gene]
        
print("supposed:0 ---",count)
print(count_threshold)
print(_max)
```
